SPOK (Subjek - Prediket - Objek - Keterangan) yang diterima :

1. Aku memasak nasi di rumah
2. Ani membeli buku di toko
3. Anak bermain bola di taman
4. Petani menanam bunga di kebun
5. Penyair menyampaikan puisi di acara

```python
class Parser:
    def __init__(self):
        self.vn = []
        self.s = ['#']
        self.state = 'i'
        self.cek = True

    def cekkalimat(self, x, y, z, a):
        self.vn = []
        self.s = ['#', 'S']
        self.state = 'p'
        self.cek = True

        while self.s[-1] != '#' and self.cek:
            top = self.s.pop()
            if top == 'S':
                if self.ceksubjek(x):
                    self.s.append('P')
                    self.vn.append('S')
                else:
                    self.cek = False
            elif top == 'P':
                if self.cekpredikat(y):
                    self.vn.append('P')
                    if z:
                        self.s.append('O')
                    elif a:
                        self.s.append('K')
                else:
                    self.cek = False
            elif top == 'O':
                if self.cekobjek(z):
                    self.vn.append('O')
                    if a:
                        self.s.append('K')
                    else:
                        self.s.append('#')
                else:
                    self.cek = False
            elif top == 'K':
                if self.cekketerangan(a):
                    self.vn.append('K')
                    self.s.append('#')
                else:
                    self.cek = False

        if self.s[-1] == '#':
            print("VALID (", ' '.join(self.vn), ")")
        else:
            print("INVALID")

    def ceksubjek(self, subjek):
      return self._cekdfa(subjek, {
          0: {'a': 1, 'p':1},
          1: {'k': 2, 'n': 2, 'e': 2},
          2: {'u': 7, 'i': 7, 'a': 3, 't': 3, 'n': 3},
          3: {'k': 7, 'a': 4, 'y': 4,},
          4: {'n': 6, 'a': 5},
          5: {'i': 6},
          6: {'i': 7, 'r': 7}
      }, 7)

    def cekpredikat(self, predikat):
        return self._cekdfa(predikat, {
            0: {'b': 1, 'm': 1},
            1: {'e': 2},
            2: {'r': 3, 'm': 3, 'n': 3},
            3: {'m': 4, 'a': 4, 'b': 4, 'y': 4},
            4: {'a': 5, 's': 5, 'e': 5, 'n': 5},
            5: {'i': 6, 'a': 6, 'l': 6, 'm': 6},
            6: {'n': 12, 'k': 12, 'i': 12, 'm': 12, 'p': 7},
            7: {'a': 8},
            8: {'i': 9},
            9: {'k': 10},
            10: {'a': 11},
            11: {'n': 12}
        }, 12)

    def cekobjek(self, objek):
        return self._cekdfa(objek, {
            0: {'n': 1, 'b': 1, 'p': 1},
            1: {'a': 2, 'u': 2, 'o': 2},
            2: {'s': 3, 'k': 3, 'l': 4, 'n': 3, 'i': 3},
            3: {'i': 5, 'u': 5, 'g': 4, 's': 4,},
            4: {'a': 5, 'i': 5},
        }, 5)



    def cekketerangan(self, keterangan):
        return self._cekdfa(keterangan, {
            0: {'d': 1},
            1: {'i': 2},
            2: {' ': 3},
            3: {'r': 4, 't': 4, 'k': 4, 'a': 4},
            4: {'u': 5, 'a': 5, 'e': 5, 'c': 5, 'o': 5},
            5: {'m': 6, 'b': 6, 'a': 6, 'k': 6},
            6: {'a': 7, 'u': 7, 'r': 7, 'o': 8},
            7: {'h': 8, 'n': 8, 'a': 8}
        }, 8)


    def _cekdfa(self, string, transitions, final_state):
        state = 0
        for char in string:
            state = transitions.get(state, {}).get(char, -1)
            if state == -1:
                return False
        return state == final_state
```

```python
if __name__ == '__main__':

    parser = Parser()

    kata = [''] * 4
    print('Input kalimat (S P) :')
    kata[0] = input()
    kata[1] = input()
    parser.cekkalimat(kata[0], kata[1], '', '')
    print('\n')

    print('Input kalimat (S P O) :')
    kata[0] = input()
    kata[1] = input()
    kata[2] = input()
    print(kata[0], kata[1], kata[2])
    parser.cekkalimat(kata[0], kata[1], kata[2], '')
    print('\n')

    print('Input kalimat (S P K) :')
    kata[0] = input()
    kata[1] = input()
    kata[2] = input()
    print(kata[0], kata[1], kata[2])
    parser.cekkalimat(kata[0], kata[1], '',  kata[2])
    print('\n')

    print('Input kalimat (S P O K) :')
    kata[0] = input()
    kata[1] = input()
    kata[2] = input()
    kata[3] = input()
    print(kata[0], kata[1], kata[2], kata[3])
    parser.cekkalimat(kata[0], kata[1], kata[2], kata[3])
```
